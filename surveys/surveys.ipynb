# Footprints

In order to add the following survey footprints

- (**DONE**) CFHTLenS
- (**DONE**) CS82
- (**DONE**) DES
- (**DONE**) HSC
- (**DONE**) KiDS
- (**DONE**) Legacy Survey DR9
- (**DONE**) RCSLenS

we used [The LAMBDA footprint library](https://github.com/nasa-lambda/cmb_footprint/).

```python
import numpy as np
import healpy as H
import pandas as pd

import sys
import os

separator = os.sep
fullpath = os.getcwd().split(separator)
source_path = separator.join(fullpath[0:-2])
sys.path.append(source_path)

from cmb_footprint import footprint

import matplotlib
import matplotlib.pyplot as plt

plt.rcParams['text.usetex'] = True
plt.rcParams['font.size'] = 18

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize = (25,18)) 

fp = footprint.SurveyStack('PLANCK-DUSTPOL', projection='mollweide', coord_bg='C', coord_plot='C', \
                           rot=[0,0], config='surveys.cfg')

fp.superimpose_survey_outline('CFHTLENS', color='green', label='CFHTLenS')
fp.superimpose_survey_outline('CS82', color='yellow', label='CS82')
fp.superimpose_survey_outline('DES-polygon', color='black', label='DES')
fp.superimpose_survey_outline('HSC', color='red', label='HSC')
fp.superimpose_survey_outline('KIDS', color='magenta', label='KiDS')
fp.superimpose_survey_outline('Legacy-polygon', color='purple', label='Legacy')
fp.superimpose_survey_outline('RCSLENS', color='blue', label='RCSLenS')

plt.savefig('plots/Survery_Footprints_C_C.pdf', dpi=300, transparent=True, bbox_inches="tight")
plt.show()
```

Let's add the unique objects:

```python
Unique = pd.read_csv('../../data-mining/Database/Files/Unique.csv')
Unique.head(), len(Unique)
```

```python
plt.figure(figsize = (25,18)) 

fp = footprint.SurveyStack('PLANCK-UT78', projection='mollweide', coord_bg='C', coord_plot='C', \
                           rot=[0,0], config='surveys.cfg', title='Unique Objects')

H.projscatter(Unique['RA'], Unique['DEC'], lonlat=True, coord='C', s=0.009, marker='o')
fp.superimpose_survey_outline('CFHTLENS', color='green', label='CFHTLenS')
fp.superimpose_survey_outline('CS82', color='yellow', label='CS82')
fp.superimpose_survey_outline('DES-polygon', color='black', label='DES')
fp.superimpose_survey_outline('HSC', color='red', label='HSC')
fp.superimpose_survey_outline('KIDS', color='magenta', label='KiDS')
fp.superimpose_survey_outline('Legacy-polygon', color='purple', label='Legacy')
fp.superimpose_survey_outline('RCSLENS', color='blue', label='RCSLenS')

plt.savefig('plots/Survery_Footprints_C_C_no_bg_with_unique_objects.pdf', transparent=True, bbox_inches="tight")
plt.show()
```
